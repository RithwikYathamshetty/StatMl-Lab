```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
r=pd.read_csv('/content/Salary_Data.csv')
print(r)
df=pd.DataFrame(r)
i=df[["YearsExperience"]]
j=df["Salary"]
print(i)
print(j)
plt.plot(i,j,'g')
```

```python
from sklearn.linear_model import LinearRegression
plt.scatter(i,j,color='red',label='data points')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title('salary vs Years of Experience')
plt.legend()
model=LinearRegression()
model.fit(i,j)
plt.plot(i,model.predict(i),color='green',label='best fit line')
plt.legend()
print("Intercept:",model.intercept_)
print("Coefficient:",model.coef_[0])
```
